```python
# Take nitrogen level from user
nitrogen = float(input("Enter soil nitrogen level (mg/kg): "))

# Check conditions
if nitrogen < 100:
    print("Recommendation: Apply HIGH nitrogen fertilizer")
    print("Cost: ₹2000 per bag")
    print("Message: Soil nitrogen is very low. Immediate fertilizer needed.")

elif nitrogen >= 100 and nitrogen <= 250:
    print("Recommendation: Apply MEDIUM nitrogen fertilizer")
    print("Cost: ₹1000 per bag")
    print("Message: Soil has moderate nitrogen. Slight supplementation required.")

else:
    print("Recommendation: No fertilizer needed")
    print("Cost: ₹0")
    print("Message: Soil is healthy and rich in nitrogen.")
```

```python
crop = input("Enter crop type (rice/wheat/sugarcane): ")
moisture = float(input("Enter soil moisture (%): "))
temp = input("Enter temperature condition (hot/moderate/cool): ")

# Decision making
if crop == "rice":
    if temp == "hot" and moisture < 60:
        print("Decision: IRRIGATE NOW")
    elif (temp == "moderate" or temp == "cool") and moisture < 50:
        print("Decision: IRRIGATE NOW")
    else:
        print("Decision: No irrigation needed")

elif crop == "wheat":
    if temp == "hot" and moisture < 40:
        print("Decision: IRRIGATE NOW")
    elif (temp == "moderate" or temp == "cool") and moisture < 30:
        print("Decision: IRRIGATE NOW")
    else:
        print("Decision: No irrigation needed")

elif crop == "sugarcane":
    if temp == "hot" and moisture < 70:
        print("Decision: IRRIGATE NOW")
    elif (temp == "moderate" or temp == "cool") and moisture < 60:
        print("Decision: IRRIGATE NOW")
    else:
        print("Decision: No irrigation needed")

else:
    print("Invalid crop type")
```

```python
pests = int(input("Enter number of pests per plant: "))
damage = float(input("Enter crop damage percentage: "))

# Severity check
if pests < 5 and damage < 10:
    print("Severity Level: LOW")
    print("Recommendation: No immediate action needed. Monitor weekly.")
    print("Cost of intervention: ₹0")

elif (pests >= 5 and pests <= 15) or (damage >= 10 and damage <= 25):
    print("Severity Level: MEDIUM")
    print("Recommendation: Monitor daily. Use mild pesticide if needed.")
    print("Cost of intervention: ₹500 per hectare")

elif pests > 15 or damage > 25:
    print("Severity Level: HIGH")
    print("Recommendation: Immediate pesticide spray required!")
    print("Cost of intervention: ₹1200 per hectare")

else:
    print("Data not in range")
```

```python
ph = float(input("Enter soil pH: "))

if ph < 5.5:
    print("Highly acidic – Add lime")

elif 5.5 <= ph < 6.5:
    print("Acidic – May need amendment")

elif 6.5 <= ph < 7.5:
    print("Neutral – Ideal soil")

elif 7.5 <= ph < 8.5:
    print("Alkaline – Add sulfur")

else:
    print("Highly alkaline – Major treatment needed")
```

```python
yield_base = 40
rain = float(input("Rainfall (mm): "))
temp = float(input("Temperature (°C): "))
soil = float(input("Soil quality (0–100): "))

# Rainfall effect
if rain > 600:
    yield_base += yield_base * 0.02 * ((rain - 600) / 10)
elif rain < 400:
    yield_base -= yield_base * 0.02 * ((400 - rain) / 10)

# Temperature effect
if temp < 25:
    yield_base -= yield_base * 0.01 * (25 - temp)
elif temp > 30:
    yield_base -= yield_base * 0.01 * (temp - 30)

# Soil quality effect
final_yield = yield_base * (soil / 100)

print("Predicted Yield:", final_yield, "quintals/hectare")
```
